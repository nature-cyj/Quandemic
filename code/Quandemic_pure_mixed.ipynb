```python
import pennylane as qml
from pennylane import numpy as np
```

```python
#GAME CODE:Ver.Pure
def game_pure(nx_qubits,ny_qubits,level,ini,state=[],game_card={}): # changed
    if ini==-1:
        dev1 = qml.device("default.qubit",wires=range(nx_qubits*ny_qubits))
        @qml.qnode(dev1)
        def ini_circuit():
            n_total = nx_qubits*ny_qubits
            coords = np.random.choice(n_total,level, replace=False)
            for coord in coords:
                qml.PauliX(wires=range(nx_qubits*ny_qubits)[coord])
            return qml.state()
        return ini_circuit()
    elif ini==-2: 
        dev2 = qml.device("default.qubit",wires=range(nx_qubits*ny_qubits))
        @qml.qnode(dev2)
        def intermediate_circuit():
            qml.AmplitudeEmbedding(state, wires=range(nx_qubits*ny_qubits))
            for swap in game_card['SWAP']:
                qml.SWAP(wires=list(map(int,swap.split(','))))
            for hospital in game_card['hospital']:
                if int(hospital) == (nx_qubits*ny_qubits-1)/2:
                    print("Welcome to the most dangerous, but efficacious hospital.")
                    qml.PauliX(wires=hospital)
                else:
                    print("Here, this is a medicine. But I'm not sure it may help you heal completely...")
                    qml.Hadamard(wires=hospital)
            return qml.state()
        return intermediate_circuit()
    elif ini == -3:
        dev3 = qml.device("default.qubit",shots=1,wires=range(nx_qubits*ny_qubits))
        @qml.qnode(dev3)
        def end_circuit():
            qml.AmplitudeEmbedding(state, wires=range(nx_qubits*ny_qubits))
            return qml.sample()
        final_pcr = end_circuit()
        if 1 not in final_pcr:
            return 'All citizens are cured!!! YOU WIN'
        else:
            return 'Virus is not disappeared. YOU LOSE'
    elif ini == -4:
        dev6 = qml.device("default.qubit",shots=1,wires=range(nx_qubits*ny_qubits))
        @qml.qnode(dev6)
        def total_survey_circuit():
            qml.AmplitudeEmbedding(state, wires=range(nx_qubits*ny_qubits))
            return qml.sample()
        dev7 = qml.device("default.qubit",wires=range(nx_qubits*ny_qubits))
        @qml.qnode(dev7)
        def after_survey(total_survey):
            print("The entire PCR test result is:",total_survey)
            qml.BasisState(total_survey, wires=range(nx_qubits*ny_qubits))
            return qml.state()
        return after_survey(total_survey_circuit())
    else:
        dev4 = qml.device("default.qubit",shots=1,wires=range(nx_qubits*ny_qubits))
        @qml.qnode(dev4)
        def measure_circuit(issample):
            qml.AmplitudeEmbedding(state, wires=range(nx_qubits*ny_qubits))
            if issample:
                return qml.sample(wires=ini)
            else:
                return qml.state()
        measure_out = measure_circuit(True)
        if measure_out == 1:
            print("The PCR test of person ",ini," is ","Positive.")
        else:
            print("The PCR test of person ",ini," is ","Negative.")
        states = measure_circuit(False)
        aft_meas = []
        j =nx_qubits*ny_qubits - 1 - ini
        for i, st in enumerate(states):
            indx = int((i % (2**(j+1)))/2**j)
            if measure_out == indx:
                aft_meas.append(st)
            else :
                aft_meas.append(0)
        dev5 = qml.device("default.qubit",wires=range(nx_qubits*ny_qubits))
        @qml.qnode(dev5)
        def re_input_circuit():
            qml.AmplitudeEmbedding(aft_meas, wires=range(nx_qubits*ny_qubits),normalize=True)
            return qml.state()
        return re_input_circuit()
            
        
```

```python
#GAME CODE:Ver.Mixed
def game_mixed(nx_qubits,ny_qubits,level,ini,state=[],game_card={}): # changed
    if ini==-1:
        dev1 = qml.device("default.mixed",wires=range(nx_qubits*ny_qubits))
        @qml.qnode(dev1)
        def ini_circuit():
            n_total = nx_qubits*ny_qubits
            coords = np.random.choice(n_total,level, replace=False)
            for coord in coords:
                qml.PauliX(wires=range(nx_qubits*ny_qubits)[coord])
            return qml.state()
        return ini_circuit()
    elif ini==-2: 
        dev2 = qml.device("default.mixed",wires=range(nx_qubits*ny_qubits)) # changed
        @qml.qnode(dev2)
        def intermediate_circuit():
            qml.QubitDensityMatrix(state, wires=range(nx_qubits*ny_qubits))
            for swap in game_card['SWAP']:
                swap_pair = list(map(int,swap.split(','))) # changed
                qml.SWAP(wires=swap_pair) # changed
                qml.PauliX(wires=swap_pair[0]) # changed
                qml.AmplitudeDamping(0.5,wires=swap_pair[0]) # changed
                qml.PauliX(wires=swap_pair[0]) # changed
                qml.PauliX(wires=swap_pair[1]) # changed
                qml.AmplitudeDamping(0.5,wires=swap_pair[1]) # changed
                qml.PauliX(wires=swap_pair[1]) # changed
            for hospital in game_card['hospital']:
                if int(hospital) == (nx_qubits*ny_qubits-1)/2:
                    print("Welcome to the most dangerous, but efficacious hospital.")
                    qml.PauliX(wires=hospital)
                else:
                    print("Here, this is a medicine. But I'm not sure it may help you heal completely...")
                    qml.Hadamard(wires=hospital)
            return qml.state()
        return intermediate_circuit()
    elif ini == -3:
        dev3 = qml.device("default.mixed",shots=1,wires=range(nx_qubits*ny_qubits)) # changed
        @qml.qnode(dev3)
        def end_circuit():
            qml.QubitDensityMatrix(state, wires=range(nx_qubits*ny_qubits)) # changed
            return qml.sample()
        final_pcr = end_circuit()
        if 1 not in final_pcr:
            return 'All citizens are cured!!! YOU WIN'
        else:
            return 'Virus is not disappeared. YOU LOSE'
    elif ini == -4:
        dev6 = qml.device("default.mixed",shots=1,wires=range(nx_qubits*ny_qubits)) # changed
        @qml.qnode(dev6)
        def total_survey_circuit():
            qml.QubitDensityMatrix(state, wires=range(nx_qubits*ny_qubits)) # changed
            return qml.sample() 
        dev7 = qml.device("default.mixed",wires=range(nx_qubits*ny_qubits)) # changed
        @qml.qnode(dev7)
        def after_survey(total_survey):
            print("The entire PCR test result is:",total_survey)
            qml.BasisState(total_survey, wires=range(nx_qubits*ny_qubits))
            return qml.state()
        return after_survey(total_survey_circuit())
    else:
        dev4 = qml.device("default.mixed",shots=1,wires=range(nx_qubits*ny_qubits)) # changed
        @qml.qnode(dev4)
        def measure_circuit(issample):
            qml.QubitDensityMatrix(state, wires=range(nx_qubits*ny_qubits)) # changed
            if issample:
                return qml.sample(wires=ini)
            else:
                return qml.state()
        measure_out = measure_circuit(True)
        if measure_out == 1:
            print("The PCR test of person ",ini," is ","Positive.")
        else:
            print("The PCR test of person ",ini," is ","Negative.")
        states = measure_circuit(False)
        aft_meas = []
        k =nx_qubits*ny_qubits - 1 - ini
        trace = 0
        # calculate the trace of post-measured density matrix
        for i, sts in enumerate(states): # changed
            indx_i = int((i % (2**(k+1)))/2**k) # changed
            if indx_i == measure_out: # changed
                trace += sts[i] # changed
        # makes the trace of post_measured density matrix equal one.
        for i, sts in enumerate(states): # changed
            aft_meas_row = [] # changed
            indx_i = int((i % (2**(k+1)))/2**k) # changed
            if indx_i == measure_out: # changed
                for j,st in enumerate(sts): # changed
                    indx_j = int((j % (2**(k+1)))/2**k) # changed
                    if measure_out == indx_j: # changed
                        aft_meas_row.append(st/trace) # changed
                    else : # changed
                        aft_meas_row.append(0) # changed
            else : # changed
                aft_meas_row += [0] * len(sts) # changed
            aft_meas.append(aft_meas_row) # changed
        dev5 = qml.device("default.mixed",wires=range(nx_qubits*ny_qubits)) # changed
        @qml.qnode(dev5)
        def re_input_circuit():
            qml.QubitDensityMatrix(aft_meas, wires=range(nx_qubits*ny_qubits)) # changed
            return qml.state()
        return re_input_circuit()
            
        
```
